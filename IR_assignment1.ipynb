### A)Information Retreval system

### Loading 5 text files 

```python
#loading into 5 documents
doc1=open('/content/365 bedtime stories by Mary Graham Bonner.txt','r')
doc2=open('/content/George Romney, by Rowley Cleeve.txt','r')
doc3=open('/content/Greuze by Harold Armitage.txt','r')
doc4=open('/content/History of the war in the Peninsula and in the south of France from the year.txt','r')
doc5=open('/content/Sir John Everett Millais by A. L. Baldry.txt','r')
```

### Preprocessing data

```python
#converting into lowercase and removing stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopword=stopwords.words('english')
doc1_procd=[i.lower() for i in doc1.read().split() if i.lower() not in stopword]
doc2_procd=[i.lower() for i in doc2.read().split() if i.lower() not in stopword]
doc3_procd=[i.lower() for i in doc3.read().split() if i.lower() not in stopword]
doc4_procd=[i.lower() for i in doc4.read().split() if i.lower() not in stopword]
doc5_procd=[i.lower() for i in doc5.read().split() if i.lower() not in stopword]
```

```python
#Special characters,punctuations and numbers removal
import re
doc1_procd=[re.sub('[^A-Za-z]+','',i) for i in doc1_procd]
print(len(doc1_procd))
doc2_procd=[re.sub('[^A-Za-z]+','',i) for i in doc2_procd]
print(len(doc2_procd))
doc3_procd=[re.sub('[^A-Za-z]+','',i) for i in doc3_procd]
print(len(doc3_procd))
doc4_procd=[re.sub('[^A-Za-z]+','',i) for i in doc4_procd]
print(len(doc4_procd))
doc5_procd=[re.sub('[^A-Za-z]+','',i) for i in doc5_procd]
print(len(doc5_procd))
```

### Forming a dictionary by itretaing through unique words

```python
#finding unique words
unq_wrds=set(doc1_procd).union(set(doc2_procd),set(doc3_procd),set(doc4_procd),set(doc5_procd))
```

```python
len(unq_wrds)
```

```python
#forming doc_dict which indicates in which document particular word is present
from collections import defaultdict
doc_dict=defaultdict(list)
for i in unq_wrds:
  if i in set(doc1_procd):
    doc_dict[i].append(1)
  else:
    doc_dict[i].append(0)
  if i in set(doc2_procd):
    doc_dict[i].append(1)
  else:
    doc_dict[i].append(0)
  if i in set(doc3_procd):
   doc_dict[i].append(1)
  else:
    doc_dict[i].append(0)
  if i in set(doc4_procd):
    doc_dict[i].append(1)
  else:
    doc_dict[i].append(0)
  if i in set(doc5_procd):
   doc_dict[i].append(1)
  else:
    doc_dict[i].append(0)
```

```python
doc_dict
```

```python
#testing againist boolean query helper function 1
import numpy as np
from numpy import array
def excute_boolean(qry):
  qry=re.sub(r'\bor\b','+',qry)
  qry=re.sub(r'\band\b','*',qry)
  qry=re.sub(r'\bnot\b','np.array([1,1,1,1,1])-',qry)
  l=qry.split()
  for i in range (len(l)):
    if l[i].isalpha():
      if doc_dict[l[i]]:
        l[i]=repr(np.array(doc_dict[l[i]]))
      else:
        l[i]=repr(np.array([0,0,0,0,0]))
  evlqry=' '.join(l)
  return (evlqry)
```

```python
#helper function 2
def redifine(l):
  for i in range (len(l)):
    if l[i]>1:
      l[i]=1
  return l
```

```python
#helper function 3
def find_docs(doc_list):
  docs=['doc1','doc2','doc3','doc4','doc5']
  for i in list(zip(docs,doc_list)):
    if i[1]==1:
      print(i[0])
```

###B) Inverted Index

```python
#forming a dictionary of index
from collections import defaultdict
invrtd_index=defaultdict(list)


for i in unq_wrds:
  if i in set(doc1_procd):
    invrtd_index[i].append(0)

  if i in set(doc2_procd):
    invrtd_index[i].append(1)

  if i in set(doc3_procd):
   invrtd_index[i].append(2)

  if i in set(doc4_procd):
    invrtd_index[i].append(3)

  if i in set(doc5_procd):
   invrtd_index[i].append(4)
 
```

```python
invrtd_index
```

```python
#query evaluation function
#helper function 1
def invinx(inp_qry):
  inp_qry=re.sub(r'\bor\b','|',inp_qry)
  inp_qry=re.sub(r'\band\b','&',inp_qry)
  inp_qry=re.sub(r'\bnot\b','{0,1,2,3,4}-',inp_qry)
  l1=inp_qry.split()
  for i in range (len(l1)):
    if l1[i].isalpha():
      if invrtd_index[l1[i]]:
        l1[i]=repr(set(invrtd_index[l1[i]]))
      else:
        l1[i]=repr(set([]))
  exp=' '.join(l1)
  return exp
```

```python
#helper function 2
def inv_match(inv_doc):
  inv_d={1:'doc1',2:'doc2',3:'doc3',4:'doc4',5:'doc5'}
  for i in set(inv_d.keys())&inv_doc:
    print(inv_d[i])
```

### Testing against different input queries
